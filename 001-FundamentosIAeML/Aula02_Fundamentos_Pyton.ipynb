#Controle de Fluxo

```python
#Condicionais (if, elif, else)
idade = 18
if idade >= 18:
  print('Você é maior de idade')
else:
  print("Você é menor de idade")
```

```python
#Laços de repetição
for i in range(5):
  print(i)

contagem = 0
while contagem < 5:
  print(contagem)
  contagem += 1
```

#Exceções


```python
#Capturando
try:
  resultado = 10/0
except ZeroDivisionError:
  print("Erro: Divisão por Zero")
```

```python
#Lançando Exceções
def dividir(a,b):
  if b == 0:
    raise ValueError("O divisor não pode ser Zero.")
  return a / b

try:
  print(dividir(10,0))
except ValueError as e:
  print(e)
```

#Decorados


```python
def decorador_saudacao(func):
  def wrapper(*args, **kwargs):
    print('Saudação')
    return func(*args, **kwargs)
  return wrapper
@decorador_saudacao
def ola(nome):
  print(f"Olá {nome}!")

ola('Felippe')
```

#Context Manager
Permite a alocação e liberação de percurso de forma eficiente, usando with

```python
class GerenciadorDeContexto:
    def __enter__(self):
        print("Entrando no contexto.")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Saindo do contexto.")

with GerenciadorDeContexto():
    print("Dentro do bloco with.")
```
