```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('face.jpg')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#Redimensionamento de imagem
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('face.jpg')

#redimensionamento
largura = 800
altura = 600
dimensoes = (largura, altura)

image_redimensionada = cv2.resize(image, dimensoes,
                                  interpolation=cv2.INTER_AREA)


#converter de BGR para RGB
image_rgb = cv2.cvtColor(image_redimensionada,
                         cv2.COLOR_BGR2RGB)

#exibir
plt.imshow(image_rgb)
#plt.axis('off')
plt.show()
```

```python
#Aplica√ß√£o de filtros e suaviza√ß√£o

import cv2
import matplotlib.pyplot as plt

imagem = cv2.imread('face.jpg')
imagem_suavizada = cv2.GaussianBlur(imagem, (15,15),0)

#converter a imagem
imagem_rgb = cv2.cvtColor(imagem_suavizada, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_rgb)
plt.axis('off')
plt.show()
```

```python
#Detec√ß√£o de bordas com Canny
import cv2
import matplotlib.pyplot as plt

imagem = cv2.imread('face.jpg')

imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#Detectando bordas
bordas = cv2.Canny(imagem_gray, 100, 200)

plt.imshow(bordas, cmap='gray')
#plt.axis('off')
plt.show()
```

```python
#Desenhando formas na imagem
import cv2
import matplotlib.pyplot as plt

imagem = cv2.imread('face.jpg')

inicio = (200, 5)
fim = (520, 320)
cor = (255, 0, 0)
espessura = 2

#Desenhando o retangulo
imagem_retangulo = cv2.rectangle(imagem.copy(), inicio, fim, cor, espessura)

#converter
imagem_retangulo_rgb = cv2.cvtColor(imagem_retangulo, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_retangulo_rgb)
plt.show()
```

```python
#Salvando
import cv2
caminho_Salvar = 'imagem_com_retangulo.jpg'
cv2.imwrite(caminho_Salvar, imagem_retangulo_rgb)
```
